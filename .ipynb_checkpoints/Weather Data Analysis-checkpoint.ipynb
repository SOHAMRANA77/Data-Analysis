# Occurrences of Clear Weather:
Calculate the frequency of occurrences when the weather condition is precisely labeled as 'Clear.'

```python
data.head(3)
```

```python
data.Weather.value_counts()
```

```python
clear_weather_count = data[data['Weather'] == 'Clear'].shape[0]
print("Occurrences of 'Clear' Weather:", clear_weather_count)
```

- data['Weather'] == 'Clear': This part of the code creates a boolean mask, which is a Series of True and False values. For each row in the 'Weather' column of the DataFrame data, it checks if the value is equal to 'Clear'. This results in a Series of True where the condition is met and False otherwise.
- data[data['Weather'] == 'Clear']: This part uses the boolean mask to filter the DataFrame data. It selects only those rows where the corresponding value in the 'Weather' column is 'Clear'.
- .shape[0]: This part accesses the .shape attribute of the filtered DataFrame, which returns a tuple representing the dimensions of the DataFrame. In this case, it's a 2D DataFrame, so the shape is represented as (number_of_rows, number_of_columns). By using [0], we're accessing the first element of this tuple, which is the number of rows in the filtered DataFrame.
- Putting it all together, data[data['Weather'] == 'Clear'].shape[0] gives the number of rows (occurrences) in the DataFrame where the 'Weather' column has the value 'Clear', effectively counting the occurrences of 'Clear' weather in the dataset.

```python
# Filtering
# data.head()
data[data.Weather == 'Clear']
```

- data.Weather: This extracts the 'Weather' column from the DataFrame data.

- data.Weather == 'Clear': This creates a boolean mask, where for each row in the 'Weather' column, it checks if the value is equal to 'Clear'. It returns True where the condition is met (i.e., the weather is 'Clear') and False otherwise.

- data[data.Weather == 'Clear']: This uses the boolean mask to filter the DataFrame data. It selects only those rows where the corresponding value in the 'Weather' column is 'Clear', effectively providing a DataFrame with rows where the weather is 'Clear'.

- So, data[data.Weather == 'Clear'] returns a DataFrame containing only the rows where the weather condition is labeled as 'Clear'.

```python
# GroupBy
# data.head()
data.groupby('Weather').get_group('Clear')
```

- data.groupby('Weather'): This groups the DataFrame data by the unique values in the 'Weather' column.
- get_group('Clear'): This retrieves the group where the 'Weather' is 'Clear'.
- The result is a DataFrame containing only the rows where the weather condition is labeled as 'Clear'. This is another way to filter the DataFrame to show only the data for when the weather is 'Clear'.

# Instances of High Wind Speed and Visibility:
Identify instances where the 'Wind Speed' is above 24 and 'Visibility' is 25.

```python
data.head(1)
```

```python
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]
```

- data['Wind Speed_km/h'] > 24: This creates a boolean mask, where for each row in the 'Wind Speed_km/h' column, it checks if the value is greater than 24. It returns True where the condition is met (i.e., wind speed is greater than 24 km/h) and False otherwise.

- data['Visibility_km'] == 25: This creates another boolean mask, where for each row in the 'Visibility_km' column, it checks if the value is equal to 25. It returns True where the condition is met (i.e., visibility is 25 km) and False otherwise.

- (data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25): This uses the & operator to combine the boolean masks from steps 1 and 2. It selects only those rows where both conditions are true, meaning the wind speed is greater than 24 km/h and the visibility is 25 km.

- So, data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)] returns a DataFrame containing only the rows where the wind speed is greater than 24 km/h and the visibility is 25 km.
